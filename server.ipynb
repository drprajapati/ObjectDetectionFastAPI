```python
from IPython.display import Image, display
```

```python
images=['apple.jpeg', 'fruits.jpeg']

for image in images:
    print(f"\nFilename: {image}")
    display(Image(filename=f"images/{image}"))
```

```python
import os
dir_name = 'results'
if not os.path.exists(dir_name):
    os.mkdir(dir_name)
```

```python
import numpy as np
import cv2
import cvlib as cv
from cvlib.object_detection import draw_bbox

def detect_and_draw(filename, model='yolov3-tiny', confidence=0.5):
    image_path = f"images/{filename}"
    image = cv2.imread(image_path)
    bbox,label, confidence = cv.detect_common_objects(image, model=model, confidence=confidence)
    print(f"========================\nImage processed: {filename}\n")
    for i, j in zip(label, confidence):
        print(f"Predicted Class Label:{i}, Predicted Class Confidence:{j}")
    
    output_image = draw_bbox(image, bbox,label, confidence)
    cv2.imwrite(f"results/{filename}", output_image)
    display(Image(filename=output_image))
```

```python
for image in images:
    detect_and_draw(image)
```
